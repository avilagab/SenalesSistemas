```python
import matplotlib.pyplot as plt
import numpy as np

#Posteriormente, se definen las características generales de las gráficas a dibujar
plt.rcParams["figure.figsize"] = [10, 5]
plt.rcParams["figure.autolayout"] = True

#La variable independiente t se tomará en el rango [0,10] y se graficará cada 0.01 segundos.
t = np.arange(-5, 5, 0.01)
x = t
# Configuración de la gráfica en particular que se va a dibujar
fig, axs = plt.subplots()
axs.set_title("Señal continua en el tiempo")
axs.axhline(0, color='0')
axs.axvline(0, color='0')
axs.plot(t, x, color='C0', linewidth=2, label='x(t) = t')
axs.set_xlabel("Tiempo")
axs.set_ylabel("Amplitud")
plt.legend(loc='upper right')
plt.show()
```

```python
#La variable independiente t se tomará en el rango [0,10] y se graficará cada 0.01 segundos.
t = np.arange(-5, 5, 0.01)
x = np.abs(t)
# Configuración de la gráfica en particular que se va a dibujar
fig, axs = plt.subplots()
axs.set_title("Señal continua en el tiempo")
axs.axhline(0, color='0')
axs.axvline(0, color='0')
axs.plot(t, x, color='C0', linewidth=2, label='x(t) = |t|')
axs.set_xlabel("Tiempo")
axs.set_ylabel("Amplitud")
plt.legend(loc='upper right')
plt.show()
```

```python
#La variable independiente t se tomará en el rango [0,10] y se graficará cada 0.01 segundos.
t = np.arange(-5, 5, 0.01)
x = np.power(t,2)
# Configuración de la gráfica en particular que se va a dibujar
fig, axs = plt.subplots()
axs.set_title("Señal continua en el tiempo")
axs.axhline(0, color='0')
axs.axvline(0, color='0')
axs.plot(t, x, color='C0', linewidth=2, label='x(t) = t^2')
axs.set_xlabel("Tiempo")
axs.set_ylabel("Amplitud")
plt.legend(loc='upper right')
plt.show()
```
